```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import talib as ta
import pynance as pn
import plotly.express as px
```

```python
ticker = 'AAPL'
start_date = '2010-01-01'
end_date = '2020-01-01'

data = yf.download(ticker, start=start_date, end=end_date)
print(data.head())
```

```python
data['SMA_20'] = ta.SMA(data['Close'],timeperiod=20)

# Calculate Relative Strength Index (RSI)
data['RSI_14'] = ta.RSI(data['Close'], timeperiod=14)

# Calculate Moving Average Convergence Divergence (MACD)
data['MACD'], data['MACD_Signal'], data['MACD_Hist'] = ta.MACD(data['Close'])

# Calculate daily returns
data['Daily_Return'] = data['Close'].pct_change()
```

```python
fig = px.line(data, x=data.index, y=['Close', 'SMA_20'], title='AAPL Stock Price with SMA')
fig.show()

# Plotting the RSI
fig = px.line(data, x=data.index, y='RSI_14', title='AAPL RSI')
fig.show()

# Plotting MACD
fig = px.line(data, x=data.index, y=['MACD', 'MACD_Signal'], title='AAPL MACD')
fig.show()
```

```python

# Display the first few rows
print(data[['Close', 'Daily_Return']].head())
fig = px.line(data, x=data.index, y='Daily_Return', title='AAPL Daily Returns')
fig.show()
```

```python
# Calculate cumulative returns
data['Cumulative_Return'] = (1 + data['Daily_Return']).cumprod() - 1

# Plotting Cumulative Returns
fig = px.line(data, x=data.index, y='Cumulative_Return', title='AAPL Cumulative Returns')
fig.show()
```
